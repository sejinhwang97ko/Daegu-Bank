# page 1~6 까지 리스트에 넣고 저장

```python
list_1 = []
for i in range(1,7): # 125개 까지 있음 1page 당 20 개
  url = 'https://www.law.go.kr/DRF/lawSearch.do?OC=sj970617&target=law&org=1160100&type=XML&page={}'.format(i)
  res = requests.get(url)
  xpars = xmltodict.parse(res.text)
  jsonDump = json.dumps(xpars)
  jsonBody = json.loads(jsonDump)
  list_1.append(jsonBody)
list_1
```

## 저장하고 불러오기

```python
with open("./law_finance.json", "w", encoding='utf-8') as json_file:
    for entry in list_1:
        json.dump(entry, json_file)
        json_file.write('\n')

with open("./law_finance.json", "r") as json_file:
    for line in json_file:
        print(line)
```

```python
def jm_key(json_data, index):
    jm = json_data[index]['조문내용']
    if '장' in jm:
        return jm
    else:
        for i in range(index, -1, -1):
            if '장' in json_data[i]['조문내용']:
                return json_data[i]['조문내용']
    return None
```

```python
Big_dict={}
Small_dict={}
for i in range(0, 10): #len(json_data_re)
    jm = json_data_re[i]['조문내용']

    if '장' in (str(jm)[2] or str(jm)[3]):
        if Small_dict:
            Big_dict.update({key_chapter: Small_dict})
            Small_dict = {}
        key_chapter = jm

    else:
        jm1 = jm
        if '항' not in json_data_re[i].keys():
            split1 = jm1.split('(')
            split_11 = jm1.split(')')
            Small_dict.setdefault(split1[0],split_11[-1])
        else:
            jm2 = json_data_re[i]['항']
            if len(jm2) == 1:
                jm3 = jm2['호']
                jm3 = jm3[0]['호내용']
                Small_dict.setdefault(jm1.split('(')[0], jm3)
            else:
                Small_dict.setdefault(jm1, jm2[0]['항내용'])

    if i == 10 - 1: # 원래 len(json_data_re)-1
        Big_dict.update({key_chapter: Small_dict})


# 장일때 {장 : }
# 조일때 {장 : {조문내용 : 항내용}}
  # 항이 없다면 -> {장 : {조문내용 : 항내용}}
  # 항이 있다면 -> 1. 항이 한개만 있다면 = 호만 존재 2. 항이 두개 이상이라면 = 항내용
```
